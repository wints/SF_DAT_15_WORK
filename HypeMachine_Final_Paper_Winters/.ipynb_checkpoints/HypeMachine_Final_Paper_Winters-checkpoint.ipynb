# Hype Machine Song Recommender

## What is Hype Machine?
**From hypem.com:**
_"Hype Machine keeps track of what music bloggers write about. We've carefully handpicked a set of 819 music blogs and then present what they discuss for easy analysis, consumption and discovery. This way, your odds of stumbling into awesome music or awesome blogs are high."_

## Why did I choose to work with Hype Machine Data?
I've been regular user of Hype Machine since mid-2009, and I still believe that it's one of the best available streaming services for discovering unique content. Spotify, Apple Music, and others are sometimes described (in a pejorative sense) as "the same 30 million songs with a search box" because the music library available to those services is largely a commodity. Certainly, both of those companies have made progress in adding value on top of the songs themselves (recommendations, social sharing, exclusive live content, etc.) but the list of songs itself remains relatively static.

Hype Machine has found a way to aggregate songs that aren't available elsewhere and has curated a distinctive music collection. _However_, Hype Machine's recommendation or personalization options are nearly nonexistent, and the primary means of discovery is the 'Popular' list, a constantly-updated feed of the entire user base's favorite songs as measured by the 'loved count' in the last 3 days ('loving' a song is Hype Machine's positive feedback mechanism). 

The 'Popular' feed offers a useful entry point into Hype Machine, but the songs are selected based entirely on the community's tastes rather than any individual's tastes. And the Hype Machine community heavily favors a handful of genres, such as:
• Dance
• EDM
• Indie
• Rap / Hip Hop

There are tens of thousands of songs in other genres that aren't easily discoverable on Hype Machine because the songs don't appear in the 'Popular' feed.

In addition, the 'Popular' feed is temporal, in that a song only remains 'Popular' while it's being actively 'loved' by users. So there are tens of thousands of songs that were once in the 'Popular' list but have since faded into obscurity. To summarize, there's a good opportunity for song recommendation.

```python
new_dict = defaultdict(list)
for page in range(0,15):
    url = "http://hypem.com/playlist/history/wints/json/"+str(page)+"/data.js"
    song_json = requests.get(url).json()
    del song_json['version']
    columns = ['mediaid', 'artist', 'title', 'dateposted' , 'sitename', 'loved_count', 'posted_count', 'time']
    for column in columns:
        new_dict[column] += [song[column] for song in song_json.values()]

songs = pd.DataFrame(new_dict)
```

There doesn't appear to be a strong relationship between the features 'loved_count', 'posted_count', and 'time' and the response 'loved', so let's think about additional features we can add using only 'songs' data. 

There's no feature that refers to a song's artist or the blog that posted the song, both of which, intuitively, could be useful features. 

Because there are hundreds of different artists and hundreds of different sites, I chose not to use the get_dummies function to create dummy variables for each artist or site, as model complexity would have increased substantially. 

Instead, I created a simple 'artist_index' and 'site_index' to measure how frequently I loved a song by a given artist or from a given site. 

The basic formulas are:

**(Total number of loved songs from an artist) / (Total number of songs from an artist on my history)**
**(Total number of loved songs from a site) / (Total number of songs from an artist on my history)**

You'll notice in the code below that I also subtracted 1 from the numerator and denominator before calculating each index value. Why? Because when evaluating the model and splitting data, I needed to account for the fact that, if I've only heard one song by a given artist or by a given site, _that data should not be available for use in making a prediction, because in the real world, if I were truly seeing a song for the first time, the prediction wouldn't have occurred yet._ 

This concept is a bit gnarly, but suffice to say: **In order for my artist and site index to reflect reality, I needed to subtract 1 from the numerator and denominator.**

```python
# code to create artist_index column
songs['artist_index_mod'] = (np.where(songs[['artist', 'loved']].groupby('artist')\
.transform('sum') > 1 , songs[['artist', 'loved']].groupby('artist')\
.transform('sum') - 1 , 0)) / (np.where(songs[['artist', 'loved']].groupby('artist').\
transform('count') > 1, songs[['artist', 'loved']].groupby('artist').\
transform('count') -1, 0))
```

```python
# code to create site_index column
songs['site_index_mod'] = (np.where(songs[['sitename', 'loved']].groupby('sitename')\
.transform('sum') > 1 , songs[['sitename', 'loved']].groupby('sitename')\
.transform('sum') - 1 , 0)) / (np.where(songs[['artist', 'loved']].groupby('artist').\
transform('count') > 1, songs[['sitename', 'loved']].groupby('sitename').\
transform('count') -1, 0))
```

```python
# need to do some quick cleaning to take care of NaN values
songs['artist_index_mod'].replace([np.inf, -np.inf], np.nan, inplace=True)
songs['site_index_mod'].replace([np.inf, -np.inf], np.nan, inplace=True)
songs['artist_index_mod'].fillna(value=0, inplace=True)
songs['site_index_mod'].fillna(value=0, inplace=True)
```

##Tags Data
My data set includes over 1000 unique tags, many of which show up on only one song (and therefore wouldn't be useful as a predictor). For the sake of model flexibility and maximizing coverage, I wanted to do my best to distill tags data into a single, wide-ranging feature rather than hard-coding specific tags as features. Here's the process I followed:
1) Run dataset containing **all** tags as predictors and 'Loved' as a reponse through a decision tree and use feature importances to determine which tags have predictive value.
2) Convert **these important tags only** to dummy columns using Pandas get_dummies
3) Create a new 'tagsum' column, summing the total number of 'important' tags appearing on each song. 
4) Keep only the 'tagsum' column on the master dataset. 

Below is a look at the relationship between the tagsum feature and 'loved' songs.

```python
sns.lmplot(x='tagsum', y='loved', data=songs_final, ci=None)
```

##Why specificity?
In a real-life scenario, the cost of listening to a song I don't like is relatively low. I don't mind having to skip through the recommender's false positives. Rather, I _don't_ want to miss out on potential songs that I might like, and so I want to avoid false negatives as much as possible. Therefore, a model with a high specificity is preferable, so long as sensitivity is in a range I find acceptable (above 50%).
